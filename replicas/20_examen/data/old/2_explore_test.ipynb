```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('./data/data_selec_test.csv')
```

```python
# Reemplazar 'NULO' y -999 por NaN
df_train = df_train.replace(['NULO', -999], np.nan)
```

```python
df_train.info()
```

```python
sns.heatmap(df_train.isnull(), cbar=False)
```

```python
fig, axs = plt.subplots(6, 5, figsize=(15, 20))
columns = df_train.select_dtypes(include=['int64', 'float64']).columns

for i, column in enumerate(columns):
    row = i // 5
    col = i % 5
    sns.histplot(df_train[column].dropna(), kde=True, ax=axs[row, col])
    axs[row, col].set_title('' + column)

plt.tight_layout()
plt.show()
```

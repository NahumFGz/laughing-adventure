```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('./data/data_selec_entre.csv')
```

```python
df_train.info()
```

```python
sns.heatmap(df_train.isnull(), cbar=False)
```

```python
for column in df_train.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure()
    sns.displot(df_train[column].dropna(), kde=True)
    plt.title('Distribucin de ' + column)
```
